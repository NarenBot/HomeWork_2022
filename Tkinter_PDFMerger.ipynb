```python
from tkinter import *
from tkinter import ttk
from tkinter import messagebox
import os
import PyPDF2

root = Tk()
root.title("PDF_Merger")
root.geometry("1920x1080+0+0")
root.config(bg="#163454")
root.state("zoomed")

pathName = StringVar()

def displayFiles():
    ls = []
    pathname = txtPath.get()
    if os.path.exists(pathname):
        os.chdir(pathname)
        rows = os.listdir()
        tv.delete(*tv.get_children())
        for row in rows:
            tv.insert("", END, values=row)
            if row.endswith(".pdf"):
                ls.append(row)
    else:
        messagebox.showerror("Error", "Please enter the valid path which exist!")

    return ls

def clearFields():
    pathName.set("")

# def getData(event):
#     pdfList = []
#     selected_row = tv.focus()
#     data = tv.item(selected_row)
#     print(data["values"])

def mergerPDF():
    pdfList = displayFiles()
    pdfCount = len(pdfList)
    if pdfCount > 1:
        merger = PyPDF2.PdfFileMerger()
        for pdf in pdfList:
            merger.append(pdf)
        output = "Final_PDF.pdf"
        with open(output, "wb") as f:
            merger.write(f)
        messagebox.showinfo("Success", "All PDF files are successfully merged in same directory under the name of Final_PDF!")
    elif pdfCount == 1:
        messagebox.showwarning("Warning", "There is only one PDF file present in current directory!")
    else:
        messagebox.showwarning("Warning", "Sorry, there are no PDF files in the current directory!")


# Input_Frame:
inputFrame = Frame(root, bg="#535c68", padx="20")
inputFrame.pack(side=TOP, fill=X)

title = Label(inputFrame, text="Browse Files:", font=("calibri", 18, "bold"), width="40", bg="#535c68", fg="white", pady="20")
title.grid(row=0, columnspan=2)

lblPath = Label(inputFrame, text="Enter the Path name:", font=("calibri", 16, "bold"), bg="#535c68", fg="white", pady="20")
lblPath.grid(row=1, column=0)
txtPath = Entry(inputFrame, textvariable=pathName, font=("calibri", 16, "italic"), fg="black", width="30")
# txtPath.insert(0, "Enter the Path name:")
txtPath.grid(row=1, column=1)

submitButton = Button(inputFrame, command=displayFiles, text="Submit", font=("calibri", 16, "bold"), bg="#16a085", fg="white",
                      bd=0, width="15")
submitButton.grid(row=2, column=0)
clearButton = Button(inputFrame, command=clearFields, text="Clear", font=("calibri", 16, "bold"), bg="#16a085", fg="white",
                     bd=0, width="15")
clearButton.grid(row=2, column=1)
mergerButton = Button(inputFrame, command=mergerPDF, text="PDF Merger", font=("calibri", 16, "bold"), bg="#16a085", fg="white",
                      bd=0, width="15")
mergerButton.grid(row=2, column=2)

# Display_Frame:
displayFrame = Frame(root, bg="#535c68")
displayFrame.pack(side=LEFT)
style = ttk.Style()
style.configure("mystyle.Treeview", font=('Calibri', 12), rowheight=50)

style.configure("mystyle.Treeview.Heading", font=('Calibri', 18, "italic"))
# displayFrame = Frame(root, bg="red")
# displayFrame.place(x=0, y=50, width=600, height=520)

tv = ttk.Treeview(displayFrame, columns=1, style="mystyle.Treeview")
tv.heading("1", text="Below are the list of files:")
tv.column("1", width="800")
tv['show'] = 'headings'
# tv.bind("<ButtonRelease-1>", getData)
tv.pack(fill=X)

# # merger_Frame:
# mergerFrame = Frame(root, bg="#535c68")
# mergerFrame.pack(side=RIGHT)
# tv1 = ttk.Treeview(mergerFrame, columns=1, style="mystyle.Treeview")
# tv1.column("1", width="50")
# tv1.pack(fill=Y)


# Path = C:\Users\Naren\Downloads\Misc

root.mainloop()

```
